### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Signal $x[n]$

```python
x = np.array([-1, -0.5, 1, 2, 2, 1, 0.5, -1])
```

### Blackman Window $w[n]$

```python
w = np.blackman(8)
w = (w * 8) / w.sum();
```

### $x[n]\cdot w[n]$

```python
x1 = x * w.transpose()
```

```python
x2 = np.zeros(16)
x2[:8] = x1
```

### Plots

```python
plt.figure(figsize=(16, 16))

# 8 samples
plt.subplot(421)
plt.stem(range(8), x)
plt.grid(True)
plt.yticks(np.arange(-1.5, 2.5, 0.5))
plt.ylabel(r'$x[n] \rightarrow$')
plt.title('a) 8 samples')
plt.tight_layout()

# 8 samples Black Window
plt.subplot(423)
plt.stem(range(8), w)
plt.grid(True)
plt.yticks(np.arange(-1.5, 3, 0.5))
plt.ylabel(r'$w[n] \rightarrow$')
plt.title('b) 8 samples Blackman window')
plt.tight_layout()

# x[n].w[n]
plt.subplot(425)
plt.stem(range(8), x1)
plt.grid(True)
plt.yticks(np.arange(-1.5, 6, 0.5))
plt.ylabel(r'$x_w[n] \rightarrow$')
plt.title(r'c) $x[n] \cdot w[n]$')
plt.tight_layout()

# x[n].w[n]+zero-padding
plt.subplot(427);
plt.stem(range(16), x2)
plt.grid(True)
plt.yticks(np.arange(-1.5, 6, 0.5))
plt.xlabel(r'$n \rightarrow$')
plt.ylabel(r'$x_w[n] \rightarrow$')
plt.title(r'd) $x[n] \cdot w[n] + zero-padding$')
plt.tight_layout()

# FFT 8 pontos
plt.subplot(222)
plt.stem(range(8), np.abs(np.fft.fft(x1)))
plt.grid(True)
plt.yticks(np.arange(-1.5, 15, 0.5))
plt.ylabel(r'$|X[k]| \rightarrow$')
plt.title('8-point FFT of c)')
plt.tight_layout()

# FFT 16 pontos
plt.subplot(224)
plt.stem(range(16), np.abs(np.fft.fft(x2)))
plt.grid(True)
plt.yticks(np.arange(-1.5, 15, 0.5))
plt.xlabel(r'$k \rightarrow$')
plt.ylabel(r'$|X[k]| \rightarrow$')
plt.title('16-point FFT of d)')
plt.tight_layout()

plt.savefig("figures/figure1_19.jpg", dpi=600, bbox_inches='tight')
```
