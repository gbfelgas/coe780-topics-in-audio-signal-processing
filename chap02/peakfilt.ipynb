```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../src')
from peakfilt import peakfilt
```

```python
fs = 48000
N = 2048
n = np.arange(N)

A0 = 0.6
A1 = 0.55
A2 = 0.55
A3 = 0.15
A4 = 0.1
A5 = 0.05

f0 = 440
f1 = 554
f2 = 660
f3 = 880
f4 = 4400
f5 = 8800

x = (A0 * np.cos(2 * np.pi * n * f0 / fs)) \
  + (A1 * np.cos(2 * np.pi * n * f1 / fs)) \
  + (A2 * np.cos(2 * np.pi * n * f2 / fs)) \
  + (A3 * np.cos(2 * np.pi * n * f3 / fs)) \
  + (A4 * np.cos(2 * np.pi * n * f4 / fs)) \
  + (A5 * np.cos(2 * np.pi * n * f5 / fs))
```

```python
plt.figure(figsize=(20,10))
plt.plot(n, x)
plt.ylabel('x[n]')
plt.xlabel('n')
plt.grid(True)
plt.show()
```

```python
fc = f0
fb = 400
Wc = 2 * fc / fs
Wb = 2 * fb / fs
G = 20

y = peakfilt(x, Wc, Wb, G)
```

```python
plt.figure(figsize=(20,10))
plt.plot(n, y)
plt.ylabel('y[n]')
plt.xlabel('n')
plt.grid(True)
plt.show()
```
